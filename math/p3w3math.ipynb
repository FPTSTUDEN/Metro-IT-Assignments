### 3. Eigen value problems

```python
import sympy as sp

def solve_matrix_problem(matrix_data, label):
    print(f"--- Solving Problem {label} ---")
    A = sp.Matrix(matrix_data)
    lam = sp.symbols('lambda') # Symbol for eigenvalue
    # set to lam to avoid confusion with Python's built-in lambda function
    
    # (i) Characteristic Equation: det(A - lambda*I) = 0
    char_poly = A.charpoly(lam)# Get the characteristic polynomial
    # Manually, we would compute det(A - lambda*I) and set it to zero, but sympy does this for us.
    char_eq = char_poly.as_expr()
    print(f"(i) Characteristic Equation: {char_eq} = 0")
    
    # (ii) Eigenvalues (found when solving above Equation)
    eigenvals = A.eigenvals()
    # eigenvals is a dictionary where keys are eigenvalues and values are their algebraic multiplicities
    # MULTIPLICITY: If an eigenvalue appears more than once as a root of the characteristic polynomial, it has a multiplicity greater than 1. This is important for understanding the structure of the eigenspace and the behavior of the matrix.
    # We just want the eigenvalues, so we take the keys of this dictionary.
    # manually, we would solve the characteristic polynomial for lambda to find these eigenvalues.
    print(f"(ii) Eigenvalues: {list(eigenvals.keys())}")
    
    # (iii) Eigenvectors
    # Returns a list of tuples: (eigenvalue, multiplicity, [eigenvectors])
    eigenvects = A.eigenvects()
    # Manually, we would substitute each eigenvalue (lambda) back into (A - lambda*I)x = 0 to find the corresponding eigenvectors.
    print("(iii) Eigenvectors:")
    for val, mult, vects in eigenvects:
        for v in vects:
            print(f"  For eigenvalue {val}: {v.T}")
        # notice that "mult" variable is not used in this code, but it indicates how many times the eigenvalue appears as a root of the characteristic polynomial, which can affect the number of linearly independent eigenvectors associated with that eigenvalue.
    print("\n")
    return eigenvects

# Problem 3 matrix
e3 = solve_matrix_problem([[5, 6], [2, 1]], "3")

# Problem 4 matrix
e4 = solve_matrix_problem([[7, -2], [1, 4]], "4")

# Problem 5 matrix
e5 = solve_matrix_problem([[5, 2], [-2, 1]], "5")
```

### 6. Like above but with var "t"

```python
# Define the symbols
x, y, z, t = sp.symbols('x y z t')

# Define the coefficient matrix A from the system:
A = sp.Matrix([
    [1, 2, -1],
    [3, 1,  t],
    [1, 1,  0]
])

print("--- Solving Problem 6 ---")

determinant = A.det()
# Manually, this would be done by calculating the determinant of the matrix 
# and setting it equal to zero.
# for 3x3 matrix, det(A) = a(ei - fh) - b(di - fg) + c(dh - eg)
print(f"1. The determinant of the matrix is: {determinant}")

# Solve the equation det(A) = 0 for t
solution = sp.solve(determinant, t)

print(f"2. Setting the determinant to zero: {determinant} = 0")
print(f"3. The value of t that allows non-trivial solutions is: t = {solution[0]}")
```

### 7. Advanced 

```python
e7 = solve_matrix_problem([[3, 2], [-3, -4]], "7")
# Model matrix M with columns are the two eigenvectors of the matrix in problem 7
v1=e7[0][2][0] # First eigenvector
v2=e7[1][2][0] # Second eigenvector
M = sp.Matrix.hstack(v1, v2)
print("--- Model Matrix M ---")
print(M)
# Show that D = M^-1 * A * M is a diagonal matrix
A = sp.Matrix([[3, 2], [-3, -4]]) # Matrix from problem 7
D = M.inv() * A * M
print("--- Diagonalized Matrix D ---")
print(D)
# Eigen values are on the diagonal of D
print(f"Eigenvalues (from diagonal of D): {D.diagonal()}")
```
